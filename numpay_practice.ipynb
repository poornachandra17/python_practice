```python
#Q15.1: Apply a function to calculate 5% interest on all account balances

interest_on_balances = bank_data[:,1] * 0.05
print(f"5% interest on all account balances:\n{interest_on_balances}")
```

```python
#Q15.2: Use np.where to categorize customers by balance ranges
balance_categories = np.where(bank_data[:,1]<5000,'Low', np.where(bank_data[:,1]<15000,'Medium','High'))
print(f"balance categories for customers:\n{balance_categories}")

```

```python
# Q15.3: Use np.select to categorize customers by credit score ranges

conditions = [
    bank_data[:,2] < 650,
    (bank_data[:,2] >= 650) & (bank_data[:,2] <= 750),
    bank_data[:,2] > 750
]

choices = ["Poor", "Average", "Excellent"]

credit_score_category = np.select(conditions, choices, default="Unknown")

print(f"credit score categories:\n{credit_score_category}")
```

```python
#Q15.4: Use np.piecewise to apply different interest rates based on balance

interest_rates = np.piecewise(bank_data[:,1], [bank_data[:,1] < 5000, (bank_data[:,1] >= 5000) & (bank_data[:,1] < 15000), bank_data[:,1] >= 15000], [0.01, 0.02, 0.03])        
print(f"interest rates based on balance:\n{interest_rates}")
```

```python
#Q15.5: Use np.apply_along_axis to calculate the sum of each row

row_sums = np.apply_along_axis(np.sum, 1, bank_data)
print(f"sum of each row:\n{row_sums}")
```
