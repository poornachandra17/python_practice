```python
# 12. Check prime
n = 7
i = 2
is_prime = True
while i < n:
    if n % i == 0:
        is_prime = False
        break
    i += 1
print( is_prime)
```

```python
# 13. Check palindrome
word = "madam"
i = 0
j = len(word) - 1
is_palindrome = True
while i < j:
    if word[i] != word[j]:
        is_palindrome = False
        break
    i += 1
    j -= 1
print( is_palindrome)
```

```python
# 14. Reverse number
num = 123
rev = 0
while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
print( rev)
```
